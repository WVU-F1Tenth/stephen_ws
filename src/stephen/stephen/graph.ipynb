```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 1. Setup the initial state
grid = np.zeros((10, 10))
plt.close('all')

fig, ax = plt.subplots()
im = ax.imshow(
    grid, 
    cmap='gray_r', 
    origin='lower', 
    interpolation='nearest', 
    vmin=0, vmax=1
)
ax.set_title("A* Search Visualization")

# 2. Define the update function
# 'step' is the current frame number provided by FuncAnimation
def update(step):
    grid[step, step] = 1
    im.set_data(grid)
    return [im]

# 3. Create the animation object
# frames=10 matches your original range(10)
# interval=200 matches your original pause(0.2)
ani = FuncAnimation(fig, update, frames=10, interval=200, repeat=False)

plt.show()
```
