```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
plt.close('all')
grid = np.zeros((10, 10))

fig, ax = plt.subplots()
im = ax.imshow(
    grid, 
    cmap='gray', 
    origin='lower', 
    interpolation='nearest', 
    vmin=0, vmax=1
)
ax.set_title("A* Search Visualization")

def update(step):
    if step == 0:
        grid.fill(0)
    grid[step, step] = 1
    im.set_data(grid)
    return [im]

ani = FuncAnimation(fig, update, frames=10, interval=200, repeat=True)

plt.show()
```
