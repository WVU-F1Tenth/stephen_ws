```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from heapq import heappush, heappop
from graph_utils import Graph
from PIL import Image
```

```python
desired_width = 100
img = Image.open('../data/images/maze2.png').transpose(Image.FLIP_TOP_BOTTOM).convert('L')
w, h = img.size
c = desired_width / w
w, h = int(w*c), int(h*c)
img = np.array(img.resize((w, h), resample=Image.Resampling.BILINEAR))
img = img >= 210
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap='gray', origin='lower')
```
